```python
import requests
import csv
from astropy.io import ascii
import numpy as np
import pandas as pd

link = "https://www.astro.keele.ac.uk/jkt/tepcat/allinfo-ascii.txt"

txt = requests.get(link).text

lines = txt.splitlines()
# Strip the leading comment marker from the header line
lines[0] = lines[0].removeprefix("#").strip()

ascii_dat = ascii.read(
    lines,
    format="fixed_width",
    delimiter=" ",
    comment=None,
)

dat = ascii_dat.to_pandas()
# Strip leading and trailing hyphens from column names
dat.columns = dat.columns.str.strip("-")

dat
```

```python
link2 = "https://www.astro.keele.ac.uk/jkt/tepcat/allinfo-csv.csv"

with requests.Session() as s:
    download = s.get(link2)

    decoded_content = download.content.decode('utf-8')

    cr = csv.reader(decoded_content.splitlines(), delimiter=',')
    my_list = list(cr)
        
dat2 = pd.DataFrame(my_list[1:], columns=my_list[0])
```

```python
dat2_nan = dat2.values.replace(-1, np.nan, inplace=True)

dat2_nan
```

```python
# select the float columns
df_num = dat2.select_dtypes(include=[np.float])
# select non-numeric columns
df_num = dat2.select_dtypes(exclude=[np.number])
```
